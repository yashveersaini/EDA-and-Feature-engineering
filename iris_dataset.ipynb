##### About Dataset:
The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

# EDA

```python
numeric_feature = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_feature = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
numeric_feature
```

```python
## Univariate analysis for numeric features
plt.figure(figsize = (12, 6))
plt.suptitle("Univariate analysis of numeric columns", color='b', fontsize=20, y=1)

for i in range(len(numeric_feature)):
    plt.subplot(2, 3, i+1)
    sns.kdeplot(x=df[numeric_feature[i]], fill=True, color='green')
    plt.tight_layout()
```

#### Conclusion:
###### * The Id column is not useful for analysis and should likely be excluded.<br>* The features SepalLengthCm, SepalWidthCm, PetalLengthCm, and PetalWidthCm can provide valuable insights, with the bimodal nature of petal dimensions hinting at species differentiation

```python
## Univariate analysis for numeric features
plt.figure(figsize=(5, 4))
plt.suptitle("Univariate analysis of categorical columns", color='b', fontsize=15, y=1)

for i in range(len(categorical_feature)):
    sns.countplot(x=df[categorical_feature[i]], color='green', hue=df[categorical_feature[i]], palette='hls')
    plt.tight_layout()
```
